```python
%pip install sqlite3
import sqlite3
```

```python
connection = sqlite3.connect('Google_Playstore.db')
c = connection.cursor()
pl_store.to_sql('Google_Playstore', connection, if_exists='replace', index=False)
```

```python
query='select "App Name", Category,Rating, max("Maximum Installs") as MAX from Google_Playstore group by Category order by MAX desc;'
table = c.execute(query).fetchall()
table
```

```python
name=[column[0] for column in c.description]
name
```

```python
table_1=pd.DataFrame(table, columns=name)
table_1
```

```python
table_1.info()
```

```python
table_1['MAX']=table_1['MAX'].astype(int)
table_1=table_1.reset_index()
table_1
```

```python
table_1.plot(kind='bar')
plt.show()
```
